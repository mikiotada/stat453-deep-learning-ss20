```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## MSE

```python
def mse(y, yhat):
    return (y-yhat)**2
```

```python
y_err1 = [mse(y=1, yhat=i) for i in np.arange(-2, 2, 0.01)]
y_err0 = [mse(y=0, yhat=i) for i in np.arange(-2, 2, 0.01)]

plt.plot(np.arange(-2, 2, 0.01), y_err1, label='y=1')
plt.plot(np.arange(-2, 2, 0.01), y_err0, label='y=0')
plt.ylabel('loss')
plt.xlabel('predicted value for y')
plt.legend()

plt.show()
```

```python
y_err1 = [mse(y=1, yhat=i) for i in np.arange(-2, 2, 0.01)]
y_err0 = [mse(y=-1, yhat=i) for i in np.arange(-2, 2, 0.01)]

plt.plot(np.arange(-2, 2, 0.01), y_err1, label='y=1')
plt.plot(np.arange(-2, 2, 0.01), y_err0, label='y=-1')
plt.ylabel('loss')
plt.xlabel('predicted value for y')
plt.legend()

plt.show()
```

## Cross-Entropy

```python
def crossentropy(y, yhat, epsilon=10**(-10)):
    return -y*(np.log(yhat + epsilon)) - ((1 - y)*(np.log(1 - yhat + epsilon)))
```

```python
y_err1 = [crossentropy(y=1, yhat=i) for i in np.arange(0, 1.01, 0.01)]
y_err0 = [crossentropy(y=0, yhat=i) for i in np.arange(0, 1.01, 0.01)]

plt.plot(np.arange(0, 1.01, 0.01), y_err1, label='y=1')
plt.plot(np.arange(0, 1.01, 0.01), y_err0, label='y=0')
plt.ylabel('loss')
plt.xlabel('predicted value for y')
plt.legend()

plt.show()
```
